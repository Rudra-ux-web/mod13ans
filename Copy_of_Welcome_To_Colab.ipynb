 Question 1: What is the difference between K-Means and Hierarchical Clustering?
Provide a use case for each.



 K Means and Hierarchical Clustering are both unsupervised machine learning algorithms used for grouping similar data points, but they differ in their approach, structure, and use cases.

K Means is a partition based clustering method. It divides the dataset into a fixed number of clusters K that is chosen before the algorithm starts. The algorithm works by randomly initializing K cluster centroids, assigning each data point to the nearest centroid, and then recalculating the centroids based on the mean of the assigned points. This process repeats until the centroids stop changing significantly. K Means is efficient and works well for large datasets, but it assumes clusters are spherical and similar in size. It also requires the number of clusters to be specified in advance.

Hierarchical Clustering, on the other hand, builds a hierarchy of clusters either by merging smaller clusters into larger ones (agglomerative approach) or by splitting a large cluster into smaller ones (divisive approach). It does not require the number of clusters to be specified at the beginning. Instead, the results are often shown using a dendrogram, a tree-like diagram that illustrates how clusters are formed. Hierarchical clustering is more flexible in capturing complex cluster shapes but is computationally more expensive and not suitable for very large datasets.

A common use case for K Means is customer segmentation in marketing, where businesses want to divide customers into a predefined number of groups based on purchasing behavior. Since the number of segments is often known in advance, K Means works well.

A common use case for Hierarchical Clustering is gene expression analysis in bioinformatics, where researchers want to explore natural groupings in the data without knowing the exact number of clusters beforehand. The dendrogram helps in understanding relationships between different genes or samples.

In summary, K Means is faster and requires a predefined number of clusters, making it suitable for large and well structured datasets, while Hierarchical Clustering provides a detailed cluster hierarchy and is useful when the number of clusters is unknown or when data relationships need to be explored.

Question 3: What are the core parameters of DBSCAN, and how do they influence the clustering process?

DBSCAN, which stands for Density Based Spatial Clustering of Applications with Noise, is a clustering algorithm that groups data points based on density rather than distance from a centroid. It is especially useful for discovering clusters of arbitrary shape and for handling noise or outliers. DBSCAN mainly depends on two core parameters: eps and min_samples, and both strongly influence how clusters are formed.

The first parameter, eps (epsilon), defines the maximum distance between two data points for them to be considered neighbors. In other words, it sets the radius of the neighborhood around a data point. If eps is too small, many points will not have enough neighbors and may be labeled as noise, resulting in too many small clusters or isolated points. If eps is too large, distinct clusters may merge together into one large cluster, reducing the quality of clustering.

The second parameter, min_samples, defines the minimum number of data points required within the eps radius for a point to be considered a core point. A core point is a point that lies in a dense region and forms the foundation of a cluster. If min_samples is set too low, the algorithm may form clusters even in sparse regions, leading to noisy or meaningless clusters. If min_samples is set too high, many points may fail to qualify as core points, and more data may be labeled as noise.

Based on these parameters, DBSCAN classifies points into three categories: core points, border points, and noise points. Core points have at least min_samples neighbors within eps distance. Border points are within eps distance of a core point but do not have enough neighbors themselves. Noise points are those that are neither core nor border points and do not belong to any cluster.

In conclusion, eps controls the size of the neighborhood around each point, while min_samples controls the minimum density required to form a cluster. Proper selection of these parameters is essential for DBSCAN to identify meaningful clusters and separate noise effectively.

Question 4: Why is feature scaling important when applying clustering algorithms like K-Means and DBSCAN?

Feature scaling is very important when applying clustering algorithms like K Means and DBSCAN because these algorithms rely heavily on distance calculations to group data points. If the features in a dataset have different scales or units, the distance metric can become biased toward features with larger values, which leads to incorrect clustering results.

In K Means clustering, the algorithm assigns data points to clusters based on the distance between the points and the cluster centroids, usually using Euclidean distance. If one feature has a much larger range than others, it will dominate the distance calculation, and the clustering will be influenced more by that feature than the others. As a result, the formed clusters may not truly represent the overall similarity between data points. Feature scaling ensures that all features contribute equally to the distance measurement.

Similarly, DBSCAN also depends on distance to determine whether points lie within a certain neighborhood radius eps. If features are not scaled, the eps value may work well for one feature but not for another. This can cause the algorithm to incorrectly identify dense regions, leading to poor cluster formation or mislabeling of points as noise.

Feature scaling techniques such as Standardization, which transforms data to have zero mean and unit variance, or Min Max scaling, which rescales features to a fixed range like 0 to 1, help bring all features to a comparable scale. This makes distance calculations more meaningful and improves the performance of clustering algorithms.

In conclusion, feature scaling is essential in clustering because it prevents features with larger scales from dominating distance calculations. By ensuring that all features have equal importance, scaling leads to more accurate and reliable clustering results when using algorithms like K Means and DBSCAN.


Question 5: What is the Elbow Method in K-Means clustering and how does it help
determine the optimal number of clusters?

The Elbow Method is a technique used in K Means clustering to help determine the optimal number of clusters for a dataset. Since K Means requires the number of clusters K to be specified in advance, the Elbow Method provides a systematic way to choose a suitable value.

In K Means, the algorithm tries to minimize the Within Cluster Sum of Squares WCSS, which measures the total distance between each data point and the centroid of its assigned cluster. A lower WCSS value indicates that the data points are closer to their cluster centroids, meaning the clusters are more compact.

In the Elbow Method, the K Means algorithm is run multiple times with different values of K, such as from 1 to 10. For each value of K, the WCSS is calculated and plotted on a graph, where the x axis represents the number of clusters and the y axis represents the WCSS value. As K increases, the WCSS decreases because data points are divided into more clusters, making them closer to their centroids.

However, after a certain point, the rate of decrease in WCSS becomes much slower. The graph typically looks like an arm, and the point where the curve bends sharply is called the elbow point. This point indicates that adding more clusters does not significantly improve clustering quality. Therefore, the value of K at the elbow is considered the optimal number of clusters.

In conclusion, the Elbow Method helps determine the best number of clusters by identifying the point where increasing K no longer provides substantial improvement in cluster compactness. It is a simple and widely used technique for selecting K in K Means clustering.







```
#Q.6

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_

# Visualize clusters and cluster centers
plt.figure()
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(centers[:, 0], centers[:, 1], marker='X', s=200)
plt.title("K-Means Clustering with Cluster Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```
#Q.7

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

# Load Wine dataset
wine = load_wine()
X = wine.data

# Apply feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train DBSCAN model
dbscan = DBSCAN(eps=1.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count number of clusters excluding noise (-1 label)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

print("Number of clusters found (excluding noise):", n_clusters)
```

```
#Question 8: Generate moon-shaped synthetic data using
#make_moons(n_samples=200, noise=0.1), apply DBSCAN, and highlight the outliers in
#the plot


import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Step 1: Generate moon-shaped synthetic data
X, y = make_moons(n_samples=200, noise=0.1, random_state=42)

# Step 2: Apply DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Step 3: Identify outliers
# In DBSCAN, outliers are labeled as -1
outliers = labels == -1
clusters = labels != -1

# Step 4: Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X[clusters, 0], X[clusters, 1], c=labels[clusters], cmap='viridis', s=50, label="Clusters")
plt.scatter(X[outliers, 0], X[outliers, 1], c='red', s=50, marker='x', label="Outliers")
plt.title("DBSCAN on Moon-shaped Data")
plt.legend()
plt.show()
```

```
#Question 9: Load the Wine dataset, reduce it to 2D using PCA, then apply
#Agglomerative Clustering and visualize the result in 2D with a scatter plot.
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering

# Step 1: Load the Wine dataset
wine = load_wine()
X = wine.data

# Step 2: Reduce to 2D using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Step 3: Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)  # Wine dataset has 3 classes
labels = agg_clustering.fit_predict(X_pca)

# Step 4: Visualize the result
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', s=50)
plt.title("Agglomerative Clustering on Wine Dataset (2D PCA)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```

```
#Question 10: You are working as a data analyst at an e-commerce company. The
#marketing team wants to segment customers based on their purchasing behavior to run
#targeted promotions. The dataset contains customer demographics and their product
#purchase history across categories.

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Create a synthetic dataset
np.random.seed(42)
df = pd.DataFrame({
    "age": np.random.randint(18, 70, 200),
    "income": np.random.randint(20000, 120000, 200),
    "frequency": np.random.randint(1, 50, 200),   # purchase frequency
    "recency": np.random.randint(1, 365, 200),    # days since last purchase
    "monetary": np.random.randint(100, 5000, 200) # total spend
})

# Step 2: Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Step 3: Reduce to 2D using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Step 4: Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Step 5: Visualize clusters
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels, cmap='viridis', s=50)
plt.title("Customer Segmentation with K-Means (Synthetic Data, 2D PCA)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```
